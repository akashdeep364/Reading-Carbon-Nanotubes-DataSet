```python
import csv
```

# Reading the raw Carbon Nanotubes Data Set


```python
# Adjusting the data in to columns , There are total 8 attributes.First we extract one single complete row.
#After extracting row we separate the semicolon separated values.
#Here i will keep the header name same as dataset and will use the same exact header name in my query.
# splitting the data with delimiter ;

file_in_list = []
with open("carbon_nanotubes.csv", 'r') as file:
    data = csv.reader(file,delimiter = '\n')  #Extracting one row
    #header = next(data)
    #print(header)
    for i in data:
        file_in_list.append(i[0].split(';'))  #Extacting ; separated values from extracted row
```

# Writing the updated data in to csv file

```python
with open('CarbonUpdatedFile.csv', 'w', newline='') as data:
    writer = csv.writer(data)
    writer.writerows(file_in_list)
```

# Creating a database in MySQL

```python
import mysql.connector as connection

try:
    #mydb = connection.connect(host="127.0.0.1",user="root", passwd="mysql",use_pure=True)
    mydb = connection.connect(host="localhost",user="root", passwd="mysql",use_pure=True)
    # check if the connection is established

    query = "Create database CarbonNanoTube;"

    cursor = mydb.cursor() #create a cursor to execute queries
    cursor.execute(query)
    print("Database Created!!")    

except Exception as e:
    mydb.close()
    print(str(e))
```

# Creating a table

```python
query = "CREATE TABLE CarbonNanoTube.NanoTubeDetails (Chiralindicen INT,Chiralindicem INT,Initialatomiccoordinateu varchar(20), Initialatomiccoordinatev varchar(20),Initialatomiccoordinatew varchar(20), Calculatedatomiccoordinatesu varchar(20), Calculatedatomiccoordinatesv varchar(20), Calculatedatomiccoordinatesw varchar(20));"
```

```python
cursor = mydb.cursor() #create a cursor to execute queries
cursor.execute(query)
print("Table Created!!")
```

# Inserting data into Table

```python
with open('CarbonUpdatedFile.csv','r') as data:
    #next(data)
    data_csv = csv.reader(data, delimiter = '\n')
    next(data_csv)
    print(data_csv)
    for j in data_csv:
        query = "INSERT INTO CarbonNanoTube.NanoTubeDetails values ({});".format(', '.join([value for value in j]))
        cursor.execute(query)
mydb.commit()
print("Values inserted")
```
